```python
# Importing pandas library as pd so we can work with data
import pandas as pd

# Helps to import the dataset

df = pd.read_csv("Customer Service_Requests_from_2010_to_Present.csv", low_memory = False)
df
```

```python
# Convert 'Created Date', 'Closed Date' column to proper datetime format & handle bad data by setting it to NaT

df['Created Date'] = pd.to_datetime(df['Created Date'], errors = 'coerce')
df['Closed Date'] = pd.to_datetime(df['Closed Date'], errors = 'coerce')
```

```python
# For "Request_Closing_Time" 

# Calculate how long each request took to close by subtracting the created date from the closed date

df['Request_Closing_Time'] = df['Closed Date'] - df['Created Date']

# Prints the Output
print(df['Request_Closing_Time'])
```

```python
# Importing numpy library for numerical operations & scipy for statistical functions

import numpy as np 
from scipy.stats import skew, kurtosis

# Select only the numeric columns from the dataframe for analysis

data = df.select_dtypes(include=[np.number])

# Replace values in datetime or timedelta columns with NaN, as they aren't needed for the numeric summary

for col in df.select_dtypes(include=['datetime64', 'timedelta64']).columns:
    data[col] = np.nan

# Calculate summary statistics like Sum, Mean, Standard Deviation, Skewness, & Kurtosis of each column

Final_Stats = pd.DataFrame({
    'Sum': data.sum(),
    'Mean': data.mean(),
    'Standard Deviation': data.std(),
    'Skewness': data.apply(lambda x: skew(x.dropna()) if x.notna().any() else np.nan),
    'Kurtosis': data.apply(lambda x: kurtosis(x.dropna()) if x.notna().any() else np.nan)
})

# Drop rows with missing values
Final_Stats.dropna()
```

```python
# Import the matplotlib library for plotting
import matplotlib.pyplot as plt

# Get the top 10 most common complaint types by counting their occurrences
Common_Complaints = df['Complaint Type'].value_counts().head(10)

# Set up the figure size for the plot
plt.figure(figsize=(12,7))

# Create a bar plot of the most common complaints
Common_Complaints.plot(kind='bar', color='lightblue')

# Adds a title to the plot
plt.title("Most Common Complaint Types")

# Label the x-axis
plt.xlabel("Number of Complaints")

print("\n\n")

# Rotate the x-axis labels to make them easier to read
plt.xticks(rotation=25)

# Display the plot
plt.show()
```

```python
# Count the number of complaints for each borough
complaints_by_borough = df['Borough'].value_counts()

# Set up the figure size for the plot
plt.figure(figsize=(12,7))

# Create a bar plot showing the number of complaints per borough
complaints_by_borough.plot(kind='bar', color='lightgreen')

# Add a title to the plot
plt.title("Number of Complaints by Borough")

# Label the y-axis (number of complaints)
plt.ylabel("Number of Complaints")

# Set the x-axis labels to be horizontal
plt.xticks(rotation=0)

print("\n\n")

# Display the plot
plt.show()
```

```python
# Calculate the average resolution time for each complaint type by grouping the data
avg_resolution = df.groupby('Complaint Type')['Request_Closing_Time'].mean()

# Convert the resolution time from timedelta to hours for easier understanding
avg_resolution_hours = avg_resolution.dt.total_seconds() / 3600

# Select the top 5 complaint types with the fastest resolution times
top_fastest = avg_resolution_hours.nsmallest(5)

# Set up the figure size for the pie chart
plt.figure(figsize=(7,7))

# Create a pie chart showing the top 5 fastest resolved complaint types with percentages
top_fastest.plot(kind='pie', autopct='%1.1f%%', startangle=140)

# Add a title to the pie chart
plt.title("Top 5 Fastest Resolved Complaint Types")

# Remove the y-axis label for clarity
plt.ylabel('')

# Display the pie chart
plt.show()
```

```python
# Compute average resolution time and convert to hours

avg_resolution = df.groupby('Complaint Type')['Request_Closing_Time'].mean()
avg_resolution_hours = avg_resolution.dt.total_seconds() / 3600

# Get top 5 slowest complaint types

top_slowest = avg_resolution_hours.nlargest(5)

# Plot as line chart

plt.figure(figsize=(11,7))
top_slowest.plot(kind='line', marker='*', color='red', markersize = 10)
plt.title("Top 5 Slowest Resolved Complaint Types")
plt.ylabel("Avg Closing Time (Hours)")
plt.xlabel("Complaint Type")

# for ploting into grid optional (plt.grid(True))

plt.xticks(rotation = 0)
plt.show()
```

```python
# Group data by 'Complaint Type' and 'Borough' and calculate the average closing time for each group
pivot = df.groupby(['Complaint Type', 'Borough'])['Request_Closing_Time'].mean().reset_index()

# data into a pivot table where 'Complaint Type' is the index, 'Borough' is the column, and 'Request_Closing_Time' is the value
pivot = pivot.pivot(index='Complaint Type', columns='Borough', values='Request_Closing_Time').fillna(0)

# Sort the pivot table by average closing time for each borough, keeping the top 10 rows
pivot = pivot.sort_values(by=pivot.columns.tolist(), ascending=False).head(10)

# Create a bar plot to visualize the average closing time for each complaint type across boroughs
pivot.plot(kind='bar', figsize=(14,7))

# Add a title to the plot
plt.title("Avg Request Closing Time by Complaint Type and Borough")

# Label the y-axis (in hours)x
plt.ylabel("Avg Time (Hours)")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=20)

print("\n\n")
# Display the plot
plt.show()
```

```python
import scipy.stats as F_stats

# Drop NaNs
filtered_df = df[['Complaint Type', 'Request_Closing_Time']].dropna()

# Convert Timedelta to hours
filtered_df['Request_Closing_Hours'] = filtered_df['Request_Closing_Time'].dt.total_seconds() / 3600

# Group by Complaint Type and extract arrays of values
groups = [group['Request_Closing_Hours'].values for _, group in filtered_df.groupby('Complaint Type')]

F_stat, P_val = F_stats.f_oneway(*groups)

print("F-statistics:", F_stat)
print("P-value:", P_val)
print("If the p-value <= 0.05, we will reject the null hypothes")
```

```python
import scipy.stats as C_stats

# Create a contingency table to count occurrences of each combination of 'Complaint Type' and 'Borough'
contingency_df = pd.crosstab(df['Complaint Type'], df['Borough'])

# Perform the Chi-Square test on the contingency table to check for independence between 'Complaint Type' and 'Borough'
Chi2_stats, P_val, dof, expected = C_stats.chi2_contingency(contingency_df)

# Print the Chi-Square statistic value
print("Chi-Square Statistic:", Chi2_stats)

# Print the p-value
print("p-value:", P_val)

print("If the p-value <= 0.05, we will reject the null hypothesis")
```
